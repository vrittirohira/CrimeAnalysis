#### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Import Dataset

```python
df2012 = pd.read_csv(r'USC2012.csv')
df2013 = pd.read_csv(r'USC2013.csv')
df2014 = pd.read_csv(r'USC2014.csv')
df2015 = pd.read_csv(r'USC2015.csv')
df2016 = pd.read_csv(r'USC2016.csv')
df2017 = pd.read_csv(r'USC2017.csv')
df2018 = pd.read_csv(r'USC2018.csv')
```

#### Find Major Crime Categories 2012-2018

```python
#----------------------To count crime records per year per category

def YearWiseCommonAttributeValuesCount(columnname, nameTODisplay):
    arg = columnname
    #Common Major Crime Categories
    cat2012 = df2012.drop_duplicates([arg])[[arg]]
    cat2013 = df2013.drop_duplicates([arg])[[arg]]
    cat2014 = df2014.drop_duplicates([arg])[[arg]]
    cat2015 = df2015.drop_duplicates([arg])[[arg]]
    cat2016 = df2016.drop_duplicates([arg])[[arg]]
    cat2017 = df2017.drop_duplicates([arg])[[arg]]
    cat2018 = df2018.drop_duplicates([arg])[[arg]]
    crime_cat = cat2012.merge(cat2013).merge(cat2014).merge(cat2015).merge(cat2016).merge(cat2017).merge(cat2018)
    cat = crime_cat[arg].tolist()
    #Retrieve those crime records in Year 2012,2013,2014,2015,2016,2017,2018
    cat_perc_2012 = df2012.loc[df2012[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2013 = df2013.loc[df2013[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2014 = df2014.loc[df2014[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2015 = df2015.loc[df2015[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2016 = df2016.loc[df2016[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2017 = df2017.loc[df2017[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2018 = df2018.loc[df2018[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    #Counts of each crime category per year
    df12_15 = pd.merge( pd.merge(pd.merge(cat_perc_2012, cat_perc_2013, on=arg), cat_perc_2014, on=arg),cat_perc_2015, on=arg)
    df12_18 = pd.merge( pd.merge(pd.merge(df12_15, cat_perc_2016, on=arg), cat_perc_2017, on=arg),cat_perc_2018, on=arg)
    df12_18.columns = [nameTODisplay, 'Year_2012','Year_2013','Year_2014','Year_2015','Year_2016','Year_2017','Year_2018']
    return df12_18
```

```python
#---------------------To run using column value eg'ccClass' (Shows Counts)
YearWiseCrimeAnalysis = YearWiseCommonAttributeValuesCount('ccClass', 'Crime Category')
print(YearWiseCrimeAnalysis)
```

```python
#------------------------------------------Analyze Percentage Difference of each common attribute category over the 2012-2018
def YearwisePercDiff(yearWiseCountData, nameToDisplay):
    yr = yearWiseCountData
    #Percentage Increase/Decrease of each attribute value over 2012-2018
    yr['2012_2013'] = round((yr['Year_2013'] - yr['Year_2012'] ) / yr['Year_2012'] * 100,2).astype(str) + " %" 
    yr['2013_2014'] = round((yr['Year_2014'] - yr['Year_2013'] ) / yr['Year_2013'] * 100,2).astype(str) + " %" 
    yr['2014_2015'] = round((yr['Year_2015'] - yr['Year_2014'] ) / yr['Year_2014'] * 100,2).astype(str) + " %" 
    yr['2015_2016'] = round((yr['Year_2016'] - yr['Year_2015'] ) / yr['Year_2015'] * 100,2).astype(str) + " %" 
    yr['2016_2017'] = round((yr['Year_2017'] - yr['Year_2016'] ) / yr['Year_2016'] * 100,2).astype(str) + " %" 
    yr['2017_2018'] = round((yr['Year_2018'] - yr['Year_2017'] ) / yr['Year_2017'] * 100,2).astype(str) + " %" 
    PercDiff = yr[[nameToDisplay,'2012_2013','2013_2014','2014_2015','2015_2016','2016_2017','2017_2018']]
    return PercDiff

YearwisePercDiff(YearWiseCrimeAnalysis, 'Crime Category')
```

```python
# ------------------------------------------Analyze cumulative percentage difference of sttribute under consideration over 2012-2018
def PercDiff_Cumulative(yearWiseCountData, nameOfAttribute):
    
    y = yearWiseCountData
    y = y.append(pd.Series(y.sum(),name='Total'))
    cum=  y.tail(1).drop(nameOfAttribute, axis=1)
    cum['2012_2013'] = round((cum['Year_2013'] - cum['Year_2012'] ) / cum['Year_2012'] * 100,2).astype(str) + " %" 
    cum['2013_2014'] = round((cum['Year_2014'] - cum['Year_2013'] ) / cum['Year_2013'] * 100,2).astype(str) + " %" 
    cum['2014_2015'] = round((cum['Year_2015'] - cum['Year_2014'] ) / cum['Year_2014'] * 100,2).astype(str) + " %" 
    cum['2015_2016'] = round((cum['Year_2016'] - cum['Year_2015'] ) / cum['Year_2015'] * 100,2).astype(str) + " %" 
    cum['2016_2017'] = round((cum['Year_2017'] - cum['Year_2016'] ) / cum['Year_2016'] * 100,2).astype(str) + " %" 
    cum['2017_2018'] = round((cum['Year_2018'] - cum['Year_2017'] ) / cum['Year_2017'] * 100,2).astype(str) + " %" 
    Cumulative_PercDiff = cum[['2012_2013','2013_2014','2014_2015','2015_2016','2016_2017','2017_2018']]
    return Cumulative_PercDiff
    
PercDiff_Cumulative(YearWiseCrimeAnalysis, 'Crime Category')
```
